```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
valores_primeira_questao = [25, 39, 18, 33, 37, 32, 26, 9, 23, 21, 34, 16, 26, 26, 18, 32, 30, 19, 31, 35, 40, 27, 35, 42, 25, 32, 21, 15, 26, 25, 33, 24]
valores_segunda_questao = [1, 6, 1, 1, 1, 3, 2, 2, 3, 2, 4, 4, 3, 13, 10, 13, 1, 2, 2, 6, 13, 5, 2, 7, 2, 1, 1, 1, 3, 2, 1, 29, 1, 3, 6, 3, 1, 2, 1, 1, 1, 2, 26, 3, 1, 1, 1, 20, 3, 1, 1, 6, 5, 1, 1, 2, 8, 1, 1, 1, 3, 12, 5, 2, 18]
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=valores_primeira_questao, orient="v", width=0.5, whis=2)
plt.title("BoxPlot dos Dados 1º Questão")
plt.ylabel("Valores")
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=valores_segunda_questao, orient="v", width=0.5)
plt.title("BoxPlot dos Dados 2º Questão")
plt.ylabel("Valores")
plt.grid(True)

plt.show()
```
