```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def equacao(x):
    return (x**2 + 2*x + 1) * np.sin(x)
```

```python
x_valores = np.linspace(-100, 100, 1000)
```

```python
y_valores = equacao(x_valores)
```

```python
df = pd.DataFrame({'x': x_valores, 'y': y_valores})
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['x'], df['y'], label='y = (x^2 + 2x + 1) * sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gr√°fico de y = (x^2 + 2x + 1) * sin(x)')
plt.grid(True)
plt.legend()
plt.show()
```
